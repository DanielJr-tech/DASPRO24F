```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('Carbon_(CO2)_Emissions_by_Country - edit_2.csv')
df
```

```python
plt.figure(figsize=(25,5))

for data in ['Kilotons of Co2', 'Metric Tons Per Capita']:
    plt.plot(df['Date'], df[data], marker='o', label=data)

plt.title("Emisi Co2 di Indonesia")
plt.ylabel("Kilotons")
plt.legend()
plt.grid(True)
plt.xticks(rotation=5)
plt.tight_layout()
plt.show
```

```python
mean_value = np.mean(df["Kilotons of Co2"])
print(mean_value)
plt.figure(figsize=(8,5))
plt.hist(df["Kilotons of Co2"], bins=15, color="skyblue", edgecolor="black", alpha=0.7, label="Hasil Mean Data")

plt.axvline(mean_value, color="red", linestyle="--", linewidth=2, label=f'Mean: {mean_value:.2f}')

plt.title("Data Distribusi Mean")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show
```

```python
plt.boxplot(df["Kilotons of Co2"], vert=False)
plt.title("Box Plot Median")
plt.xlabel("Value")
plt.show
```

```python
mean_KilotonsofCo2 = df["Kilotons of Co2"].mean()
std_KilotonsofCo2 = df["Kilotons of Co2"].std()
plt.errorbar(["Kilotons of Co2"], [mean_KilotonsofCo2], yerr=[std_KilotonsofCo2], fmt="o", capsize=5, label="Mean +- Std Dev")

plt.title("Mean dan Standar Deviasi Kilotons Co2")
plt.ylabel("Value")
plt.show()
```

```python
x = np.arange(len(df)).reshape(-1,1)
y = df["Kilotons of Co2"].values
model = LinearRegression()
model.fit(x,y)

y_pred = model.predict(x)
plt.scatter(df["Date"], y, label="Data Asli", color="gold", marker="o")

plt.plot(df["Date"], y, label="Garis Regresi", color="black", linestyle="--")

plt.title("Garis Regresi Kilotons Co2")
plt.xlabel("Date")
plt.ylabel("Kilotons of Co2 Value")
plt.legend()

plt.xticks(rotation=45)
plt.show
```
